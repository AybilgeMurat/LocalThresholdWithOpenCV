```python
import numpy as np
import cv2 
from matplotlib import pyplot as plt
from image_slicer import slice
```

```python
img=cv2.imread("sokak2.jpg",0) #zrafa ve defter iin deitirelecek
ret1,th1=cv2.threshold(img.copy(),127, 255, cv2.THRESH_BINARY)
ret2,th2=cv2.threshold(img.copy(),0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

print("binary",ret1)
print("otsu",ret2)

titles=["Orijinal","Binary","Otsu"]
images=[img,th1,th2]

plt.figure(figsize=(20,20))
for i in range(3):
    plt.subplot(3,3,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
